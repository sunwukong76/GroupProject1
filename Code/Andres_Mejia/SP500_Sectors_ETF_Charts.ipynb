```python
import pandas as pd
from pathlib import Path
import datetime
import hvplot.pandas
```

```python
# S&P500 

file_path_E = Path("../../Resources/Data/SPY_Sectors_Ebola_AM.csv")
file_path_H = Path("../../Resources/Data/SPY_Sectors_H1N1_AM.csv")
file_path_S = Path("../../Resources/Data/SPY_Sectors_SARS_AM.csv")
file_path_Z = Path("../../Resources/Data/SPY_Sectors_Zika_AM.csv")

EBOLA_spy_data = pd.read_csv(file_path_E)
H1N1_spy_data = pd.read_csv(file_path_H)
SARS_spy_data = pd.read_csv(file_path_S)
ZIKA_spy_data = pd.read_csv(file_path_Z)

EBOLA_spy_data["Date"]=pd.to_datetime(EBOLA_spy_data["Date"])
ZIKA_spy_data["Date"]=pd.to_datetime(ZIKA_spy_data["Date"])
H1N1_spy_data["Date"]=pd.to_datetime(H1N1_spy_data["Date"])
SARS_spy_data["Date"]=pd.to_datetime(SARS_spy_data["Date"])

# Gold & T-Bonds ETFs

# Gold
file_path_GOLD_E = Path("../../Resources/Data/GOLD_Ebola_AM.csv")
file_path_GOLD_H = Path("../../Resources/Data/GOLD_H1N1_AM.csv")
file_path_GOLD_S = Path("../../Resources/Data/GOLD_SARS_AM.csv")
file_path_GOLD_Z = Path("../../Resources/Data/GOLD_Zika_AM.csv")

EBOLA_GOLD_data = pd.read_csv(file_path_GOLD_E)
H1N1_GOLD_data = pd.read_csv(file_path_GOLD_H)
SARS_GOLD_data = pd.read_csv(file_path_GOLD_S)
ZIKA_GOLD_data = pd.read_csv(file_path_GOLD_Z)

EBOLA_GOLD_data["Date"]=pd.to_datetime(EBOLA_GOLD_data["Date"])
ZIKA_GOLD_data["Date"]=pd.to_datetime(ZIKA_GOLD_data["Date"])
H1N1_GOLD_data["Date"]=pd.to_datetime(H1N1_GOLD_data["Date"])
SARS_GOLD_data["Date"]=pd.to_datetime(SARS_GOLD_data["Date"])

# TLT
file_path_TLT_E = Path("../../Resources/Data/TLT_Ebola_AM.csv")
file_path_TLT_H = Path("../../Resources/Data/TLT_H1N1_AM.csv")
file_path_TLT_S = Path("../../Resources/Data/TLT_SARS_AM.csv")
file_path_TLT_Z = Path("../../Resources/Data/TLT_Zika_AM.csv")

EBOLA_TLT_data = pd.read_csv(file_path_TLT_E)
H1N1_TLT_data = pd.read_csv(file_path_TLT_H)
SARS_TLT_data = pd.read_csv(file_path_TLT_S)
ZIKA_TLT_data = pd.read_csv(file_path_TLT_Z)

EBOLA_TLT_data["Date"]=pd.to_datetime(EBOLA_TLT_data["Date"])
ZIKA_TLT_data["Date"]=pd.to_datetime(ZIKA_TLT_data["Date"])
H1N1_TLT_data["Date"]=pd.to_datetime(H1N1_TLT_data["Date"])
SARS_TLT_data["Date"]=pd.to_datetime(SARS_TLT_data["Date"])
```

```python
# Add a virus name column to each of the SPY dataframes
SARS_spy_data['Disease']='SARS'
H1N1_spy_data['Disease']='H1N1'
EBOLA_spy_data['Disease']='EBOLA'
ZIKA_spy_data['Disease']='ZIKA'

# Add ETF name column to each of the ETF dataframes
# Gold
SARS_GOLD_data['Disease']='SARS'
H1N1_GOLD_data['Disease']='H1N1'
EBOLA_GOLD_data['Disease']='EBOLA'
ZIKA_GOLD_data['Disease']='ZIKA'
SARS_TLT_data['Disease']='SARS'
H1N1_TLT_data['Disease']='H1N1'
EBOLA_TLT_data['Disease']='EBOLA'
ZIKA_TLT_data['Disease']='ZIKA'

#TLT
SARS_GOLD_data['ETF']='GOLD'
H1N1_GOLD_data['ETF']='GOLD'
EBOLA_GOLD_data['ETF']='GOLD'
ZIKA_GOLD_data['ETF']='GOLD'
SARS_TLT_data['ETF']='TLT'
H1N1_TLT_data['ETF']='TLT'
EBOLA_TLT_data['ETF']='TLT'
ZIKA_TLT_data['ETF']='TLT'
```

```python
# Concatenate the dataframes together
dis_concat = pd.concat([SARS_spy_data,EBOLA_spy_data,H1N1_spy_data,ZIKA_spy_data], axis=0, join="inner", sort=False)
etf_concat = pd.concat([SARS_GOLD_data,EBOLA_GOLD_data,H1N1_GOLD_data,ZIKA_GOLD_data, SARS_TLT_data,EBOLA_TLT_data,H1N1_TLT_data,ZIKA_TLT_data], axis=0, join="inner", sort=False)
```

```python
dis_grp=dis_concat.set_index(['Disease','Date'])
etf_grp=etf_concat.set_index(['ETF', 'Disease'])
etf_grp.dropna(inplace=True)
```
