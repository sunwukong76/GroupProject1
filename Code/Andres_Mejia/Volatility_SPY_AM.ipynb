```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
# Reading CandleSticks & Volume Dtata

candlestick_csv = Path("SPY_Ebola_AM.csv")
candlestick_data = pd.read_csv(candlestick_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
candlestick_data.head(5)
```

```python
candlestick_data[['Open', 'High', 'Low', 'Close']].plot(title = 'Ebola')
```

```python
candlestick_data["O<->C"]=candlestick_data[['Open', 'Close']].max(axis=1)-candlestick_data[['Open', 'Close']].min(axis=1)


#joined_data_cols['Bu<->Be'] = joined_data_cols[['Bullish', 'Bearish']].max(axis=1)-joined_data_cols[['Bullish', 'Bearish']].min(axis=1)
```

```python
candlestick_data.head()
```

```python
Total_OC = candlestick_data['O<->C'].sum()
print(Total_OC)
candlestick_data['%OC'] = candlestick_data['O<->C']/Total_OC
candlestick_data.head()
```

```python
Total_Volume = candlestick_data['Volume'].sum()
print(Total_Volume)
candlestick_data['%Vol'] = candlestick_data['Volume']/Total_Volume
candlestick_data.head()
```

```python
Volatility = candlestick_data['Volume'].sum()
print(Total_Volume)
candlestick_data['Volatility'] = candlestick_data['%OC']+candlestick_data['%OC']
Volatility = candlestick_data['Volatility'].sum()
candlestick_data.head()
```
