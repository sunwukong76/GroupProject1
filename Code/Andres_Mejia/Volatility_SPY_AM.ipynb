```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

```python
# Reading the csv Files with Diseases Data

file_path_E = Path("../../Resources/Data/SPY_Sectors_Ebola_AM.csv")
file_path_H = Path("../../Resources/Data/SPY_Sectors_H1N1_AM.csv")
file_path_S = Path("../../Resources/Data/SPY_Sectors_SARS_AM.csv")
file_path_Z = Path("../../Resources/Data/SPY_Sectors_Zika_AM.csv")

EBOLA_spy_data = pd.read_csv(file_path_E)
H1N1_spy_data = pd.read_csv(file_path_H)
SARS_spy_data = pd.read_csv(file_path_S)
ZIKA_spy_data = pd.read_csv(file_path_Z)

EBOLA_spy_data["Date"]=pd.to_datetime(EBOLA_spy_data["Date"])
ZIKA_spy_data["Date"]=pd.to_datetime(ZIKA_spy_data["Date"])
H1N1_spy_data["Date"]=pd.to_datetime(H1N1_spy_data["Date"])
SARS_spy_data["Date"]=pd.to_datetime(SARS_spy_data["Date"])

# Add a virus name column to each of these dataframes

SARS_spy_data['Disease']='SARS'
H1N1_spy_data['Disease']='H1N1'
EBOLA_spy_data['Disease']='EBOLA'
ZIKA_spy_data['Disease']='ZIKA'

# Concatenate the dataframes together

SP500_data = pd.concat([SARS_spy_data,EBOLA_spy_data,H1N1_spy_data,ZIKA_spy_data], axis=0, join="inner", sort=False).set_index(['Disease','Date'])

SP500_data.head()
```

```python
# DataFrames with %Change per Disease

SP500_Sectors_ZIKA = SP500_data.loc['ZIKA',('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume')].pct_change().abs()
SP500_Sectors_H1N1 = SP500_data.loc['H1N1',('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume')].pct_change().abs()
SP500_Sectors_SARS = SP500_data.loc['SARS',('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume')].pct_change().abs()
SP500_Sectors_EBOLA = SP500_data.loc['EBOLA',('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume')].pct_change().abs()

SP500_Sectors_EBOLA.head(10)
```

```python
#Calculate Limits-Points per Disease to classify Volatility

#ZIKA
min_ZIKA_data = pd.DataFrame(SP500_Sectors_ZIKA.min()).transpose()
max_ZIKA_data = pd.DataFrame(SP500_Sectors_ZIKA.max()).transpose()
mean_ZIKA_data = pd.DataFrame(SP500_Sectors_ZIKA.mean()).transpose()
aux_df_ZIKA = pd.concat([min_ZIKA_data, max_ZIKA_data, mean_ZIKA_data], axis=0, join="inner", sort=False)
aux_df_ZIKA ['min/max/mean'] = ['min','max','mean']
aux_df_ZIKA ['Disease'] = 'ZIKA'

#H1N1
min_H1N1_data = pd.DataFrame(SP500_Sectors_H1N1.min()).transpose()
max_H1N1_data = pd.DataFrame(SP500_Sectors_H1N1.max()).transpose()
mean_H1N1_data = pd.DataFrame(SP500_Sectors_H1N1.mean()).transpose()
aux_df_H1N1 = pd.concat([min_H1N1_data, max_H1N1_data, mean_H1N1_data], axis=0, join="inner", sort=False)
aux_df_H1N1 ['min/max/mean'] = ['min','max','mean']
aux_df_H1N1 ['Disease'] = 'H1N1'

#SARS
min_SARS_data = pd.DataFrame(SP500_Sectors_SARS.min()).transpose()
max_SARS_data = pd.DataFrame(SP500_Sectors_SARS.max()).transpose()
mean_SARS_data = pd.DataFrame(SP500_Sectors_SARS.mean()).transpose()
aux_df_SARS = pd.concat([min_SARS_data, max_SARS_data, mean_SARS_data], axis=0, join="inner", sort=False)
aux_df_SARS ['min/max/mean'] = ['min','max','mean']
aux_df_SARS ['Disease'] = 'SARS'

#EBOLA
min_EBOLA_data = pd.DataFrame(SP500_Sectors_EBOLA.min()).transpose()
max_EBOLA_data = pd.DataFrame(SP500_Sectors_EBOLA.max()).transpose()
mean_EBOLA_data = pd.DataFrame(SP500_Sectors_EBOLA.mean()).transpose()
aux_df_EBOLA = pd.concat([min_EBOLA_data, max_EBOLA_data, mean_EBOLA_data], axis=0, join="inner", sort=False)
aux_df_EBOLA ['min/max/mean'] = ['min','max','mean']
aux_df_EBOLA ['Disease'] = 'EBOLA'

SP500_Sectors_Change = pd.concat([aux_df_ZIKA, aux_df_H1N1, aux_df_SARS, aux_df_EBOLA], axis=0, join="inner", sort=False).set_index(['Disease','min/max/mean'])
```

```python
bins_dict = {'Ticker':'bins'}
for disease in ('ZIKA', 'H1N1', 'SARS', 'EBOLA'):
    for ticker in ('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume'):
        bins_dict[f'{ticker}_{disease}'] = [SP500_Sectors_Change.loc[disease,'min'][ticker], SP500_Sectors_Change.loc[disease,'mean'][ticker], SP500_Sectors_Change.loc[disease,'max'][ticker]]
```

```python
group_names = ['L', 'H']

# Volatility for Zika
SP500_Sectors_ZIKA.dropna(inplace=True)
for ticker in ('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume'):
    SP500_Sectors_ZIKA[f'V_{ticker}']=pd.cut(SP500_Sectors_ZIKA[ticker], bins_dict[f'{ticker}_ZIKA'],labels=group_names, include_lowest=True)
SP500_Volatility_ZIKA = SP500_Sectors_ZIKA[['V_SP500_Close', 'V_SP500_Volume', 'V_XLY_Close', 'V_XLY_Volume', 'V_XLP_Close', 'V_XLP_Volume', 'V_XLE_Close', 'V_XLE_Volume', 'V_XLF_Close', 'V_XLF_Volume', 'V_XLV_Close', 'V_XLV_Volume', 'V_XLI_Close', 'V_XLI_Volume', 'V_XLB_Close', 'V_XLB_Volume', 'V_XLK_Close', 'V_XLK_Volume', 'V_XLU_Close', 'V_XLU_Volume']]
#SP500_Volatility_ZIKA.head()

# Volatility for H1N1
SP500_Sectors_H1N1.dropna(inplace=True)
for ticker in ('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume'):
    SP500_Sectors_H1N1[f'V_{ticker}']=pd.cut(SP500_Sectors_H1N1[ticker], bins_dict[f'{ticker}_H1N1'],labels=group_names, include_lowest=True)
SP500_Volatility_H1N1 = SP500_Sectors_H1N1[['V_SP500_Close', 'V_SP500_Volume', 'V_XLY_Close', 'V_XLY_Volume', 'V_XLP_Close', 'V_XLP_Volume', 'V_XLE_Close', 'V_XLE_Volume', 'V_XLF_Close', 'V_XLF_Volume', 'V_XLV_Close', 'V_XLV_Volume', 'V_XLI_Close', 'V_XLI_Volume', 'V_XLB_Close', 'V_XLB_Volume', 'V_XLK_Close', 'V_XLK_Volume', 'V_XLU_Close', 'V_XLU_Volume']]
#SP500_Volatility_H1N1.head()

# Volatility for SARS
SP500_Sectors_SARS.dropna(inplace=True)
for ticker in ('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume'):
    SP500_Sectors_SARS[f'V_{ticker}']=pd.cut(SP500_Sectors_SARS[ticker], bins_dict[f'{ticker}_SARS'],labels=group_names, include_lowest=True)
SP500_Volatility_SARS = SP500_Sectors_SARS[['V_SP500_Close', 'V_SP500_Volume', 'V_XLY_Close', 'V_XLY_Volume', 'V_XLP_Close', 'V_XLP_Volume', 'V_XLE_Close', 'V_XLE_Volume', 'V_XLF_Close', 'V_XLF_Volume', 'V_XLV_Close', 'V_XLV_Volume', 'V_XLI_Close', 'V_XLI_Volume', 'V_XLB_Close', 'V_XLB_Volume', 'V_XLK_Close', 'V_XLK_Volume', 'V_XLU_Close', 'V_XLU_Volume']]
SP500_Volatility_SARS.head()

# Volatility for EBOLA
SP500_Sectors_EBOLA.dropna(inplace=True)
for ticker in ('SP500_Close', 'SP500_Volume', 'XLY_Close', 'XLY_Volume', 'XLP_Close', 'XLP_Volume', 'XLE_Close', 'XLE_Volume', 'XLF_Close', 'XLF_Volume', 'XLV_Close', 'XLV_Volume', 'XLI_Close', 'XLI_Volume', 'XLB_Close', 'XLB_Volume', 'XLK_Close', 'XLK_Volume', 'XLU_Close', 'XLU_Volume'):
    SP500_Sectors_EBOLA[f'V_{ticker}']=pd.cut(SP500_Sectors_EBOLA[ticker], bins_dict[f'{ticker}_EBOLA'],labels=group_names, include_lowest=True)
SP500_Volatility_EBOLA = SP500_Sectors_EBOLA[['V_SP500_Close', 'V_SP500_Volume', 'V_XLY_Close', 'V_XLY_Volume', 'V_XLP_Close', 'V_XLP_Volume', 'V_XLE_Close', 'V_XLE_Volume', 'V_XLF_Close', 'V_XLF_Volume', 'V_XLV_Close', 'V_XLV_Volume', 'V_XLI_Close', 'V_XLI_Volume', 'V_XLB_Close', 'V_XLB_Volume', 'V_XLK_Close', 'V_XLK_Volume', 'V_XLU_Close', 'V_XLU_Volume']]
SP500_Volatility_EBOLA.head()
```
