```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# Reading csv SP500 Sectors Data for Zika, H1N1, SARS and Ebola

zika_csv = Path("SPY_Sectors_Zika_AM.csv")
zika_data = pd.read_csv(zika_csv)#, index_col='Date', parse_dates=True, infer_datetime_format=True)
h1n1_csv = Path("SPY_Sectors_H1N1_AM.csv")
h1n1_data = pd.read_csv(h1n1_csv)#, index_col='Date', parse_dates=True, infer_datetime_format=True)
sars_csv = Path("SPY_Sectors_SARS_AM.csv")
sars_data = pd.read_csv(sars_csv)#, index_col='Date', parse_dates=True, infer_datetime_format=True)
ebola_csv = Path("SPY_sectors_ebola_AM.csv")
ebola_data = pd.read_csv(ebola_csv)#, index_col='Date', parse_dates=True, infer_datetime_format=True)
```

```python
plot_zika = zika_data[['Date', 'XLY_Close', 'XLP_Close', 'XLE_Close', 'XLF_Close', 'XLV_Close', 'XLI_Close', 'XLB_Close', 'XLK_Close', 'XLU_Close']]
plot_zika.plot(x = 'Date', xlim=(0,670), figsize=(20,10), title='Zika S&P500 Sectors')
```

```python
plot_h1n1 = h1n1_data[['Date', 'XLY_Close', 'XLP_Close', 'XLE_Close', 'XLF_Close', 'XLV_Close', 'XLI_Close', 'XLB_Close', 'XLK_Close', 'XLU_Close']]
plot_h1n1.plot(x = 'Date', xlim=(0,370), figsize=(20,10), title='H1N1 S&P500 Sectors')
```

```python
plot_sars = sars_data[['Date', 'XLY_Close', 'XLP_Close', 'XLE_Close', 'XLF_Close', 'XLV_Close', 'XLI_Close', 'XLB_Close', 'XLK_Close', 'XLU_Close']]
plot_sars.plot(x = 'Date', xlim=(0,90), figsize=(20,10), title='SARS S&P500 Sectors')
```

```python
plot_ebola = ebola_data[['Date', 'XLY_Close', 'XLP_Close', 'XLE_Close', 'XLF_Close', 'XLV_Close', 'XLI_Close', 'XLB_Close', 'XLK_Close', 'XLU_Close']]
plot_ebola.plot(x = 'Date', xlim=(0,580), figsize=(20,10), title='Ebola S&P500 Sectors')
```
